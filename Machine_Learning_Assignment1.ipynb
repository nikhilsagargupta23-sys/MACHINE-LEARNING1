**Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?**
**Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.**

1. **Overfitting**
**Definition:**
Overfitting happens when a model learns the training data too well, including noise and unnecessary patterns. As a result, the model performs very well on training data but poorly on new/unseen data.
Causes:
Model is too complex
Too many parameters
Insufficient or noisy training data
Training for too many epochs
How to Detect Overfitting:
High training accuracy but low test accuracy
Large gap between training loss and validation loss
Poor performance on cross-validation sets
How to Prevent Overfitting:
Cross-validation (e.g., k-fold CV)
Regularization (L1, L2)
Dropout (for neural networks)
Early stopping
Pruning decision trees
Using simpler models
Add more training data

2. **Underfitting**
**Definition**:
Underfitting occurs when a model is too simple to learn the underlying patterns in the data. It performs poorly on both training and test data.
Causes:
Model is too simple
Too few parameters
Not enough training time
Incorrect features or missing important features
How to Detect Underfitting:
Low training accuracy and low test accuracy
High bias (model makes overly simple assumptions)
Training and validation loss are both high
How to Prevent Underfitting:
Use a more complex model
Increase training time/epochs
Add more relevant features
Reduce regularization strength
Use advanced algorithms (e.g., Random Forest, Gradient Boosting)

  **3. Bias–Variance Tradeoff**
     **Overfitting** = high variance + low bias
     (Model is too complex and reacts too much to training data.)
     Underfitting = high bias + low variance
     (Model is too simple and unable to learn patterns.)
   A good model finds a balance between bias (simplicity) and variance (complexity).


**Q3. How would you handle missing values in dataset? explain altest three methods with examples. Hint Consider deletion mean/median imputation, and predictive modeling.**

1. **Deletion Method:**
In this method, the rows or columns containing missing values are removed from the dataset. This is useful when the number of missing values is very small. For example, if we have a table of Age and Salary and only one row has Salary = NaN, we can remove that entire row. This method works only when deleting a few rows does not affect the overall data quality.

2. **Mean/Median Imputation:**
In this method, the missing values are replaced with the mean or median of that particular column. Mean imputation is used when the data has no extreme outliers, while median imputation is better when the column contains outliers.
Example: If the Salary column has values 30000, NaN, and 40000, then the mean is 35000. So we replace the missing value with 35000. Similarly, if the data contains outliers, we use the median value to fill the missing entry.

3. **Predictive Modeling:**
This is an advanced method where machine learning models are used to predict the missing values. The column with missing values is treated as the target, and the remaining columns are used as input features. Algorithms like Linear Regression, KNN, or Random Forest can be used. For example, if Salary is missing but Age and Experience are available, a model can be trained using the existing rows to predict the missing Salary value.

**Question4 : What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).**

**Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models**

1. **Random Oversampling**
**Theory:**
In random oversampling, we increase the size of the minority class by repeating (duplicating) some of its samples. This helps the model learn more about the small class.
Practical Example:
If Class 0 has 900 samples and Class 1 has only 100 samples, we duplicate the minority samples until both classes have equal numbers (for example, 900 vs 900). This improves the model’s ability to identify the minority class cases.

2. **SMOTE (Synthetic Minority Oversampling Technique)**
**Theory:**
SMOTE is an advanced oversampling method. Instead of simply copying existing minority samples, it creates new synthetic samples by combining and interpolating between nearby minority data points. This makes the dataset more realistic and better balanced.
Practical Example:
If Class 1 has only 100 samples, SMOTE generates new samples similar to them, such as creating a new data point between two existing minority class points (not an exact copy). This increases the size of the minority class in a more natural way than simple duplication.

**Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.**

**Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.**

 - Feature scaling is important in Machine Learning because different features may have different ranges. For example, Age may be between 20–60, while Salary may be between 20,000–1,00,000. If we use such unscaled data, models may give more importance to the feature with the larger value. Scaling helps all features contribute equally.
It is especially important for distance-based algorithms like KNN, K-Means, SVM, and for algorithms that use gradient descent, because these methods are sensitive to the magnitude of values.

1. **Min-Max Scaling (Normalization)**
**Theory:**
Min-Max scaling converts the values of a feature into a fixed range, usually between 0 and 1.
Formula:
( X – min ) / ( max – min )
Example:
If Age = 40, min = 20, max = 60,
scaled value = (40 – 20) / (60 – 20) = 0.5.
When to use:
When you need values in the 0–1 range
Works well for KNN, K-Means, Neural Networks
Not suitable when the data has too many outliers

2. **Standardization (Z-score Scaling)**
**Theory:**
Standardization converts the data so that it has a mean of 0 and standard deviation of 1.
Formula:
( X – mean ) / standard deviation
Example:
If Height = 170, mean = 160, std = 10,
scaled value = (170 – 160) / 10 = 1.0.
When to use:
Works well even when data has outliers
Useful for SVM, Logistic Regression, Linear Regression, PCA
Values are not restricted to 0–1

**Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**

**Hint: Consider categorical variables with ordinal vs. nominal relationships.**

- Label Encoding and One-Hot Encoding are two techniques used to convert categorical (non-numerical) data into numerical form so that machine learning models can understand them. Both serve different purposes based on the type of category.

1. **Label Encoding**
**Definition:**
Label Encoding assigns a unique integer number to each category in a column.
Example:
Colors = {Red, Blue, Green}
After encoding: Red = 0, Blue = 1, Green = 2
When to use:
When the categorical variable has a natural order (ordinal data)
Examples:
• Low, Medium, High
• Small, Medium, Large
Models like decision trees can handle label-encoded data well.
Limitation:
It creates an artificial order when the categories do not actually have any ranking.

2. **One-Hot Encoding**
**Definition:**
One-Hot Encoding creates a new column for each category and marks the presence with 1 and absence with 0.
Example:
Colors = Red, Blue, Green
Becomes:
Red → [1, 0, 0]
Blue → [0, 1, 0]
Green → [0, 0, 1]
When to use:
When the categorical variable has no natural order (nominal data)
Examples:
• Gender (Male, Female)
• City (Mumbai, Delhi, Kolkata)
• Colors (Red, Blue, Green)
Limitation:
If the category has many unique values, it creates too many new columns (high dimensionality).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for Delhi-Mumbai route
delhi_mumbai_flights = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Calculate average price by airline
airline_prices = delhi_mumbai_flights.groupby('airline')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=airline_prices.index, y=airline_prices.values, palette='coolwarm')
plt.title('Average Price by Airline (Delhi–Mumbai)')
plt.xticks(rotation=45)
plt.ylabel('Average Price')
plt.show()
```
